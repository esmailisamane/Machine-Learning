```python
colourSpaceimg = plt.imread(r'ColourSpace.png')
plt.imshow(colourSpaceimg)
```

```python
def f_rgb_to_hsv(r, g, b, scaleFactor):
    r, g, b = r/255.0, g/255.0, b/255.0
    
    cmax = max(r,g,b)
    cmin = min(r,g,b)
    diff = cmax - cmin
    if cmax == cmin:
        h = 0
    elif cmax == r:
            h = (60 *((g -b)/ diff)+0) %360
    elif cmax == g:
            h = (60 *((b -r)/ diff)+120) %360
    elif cmax == b:
            h = (60 *((r -g)/ diff)+240) %360
    if h<0:
        h = h + 360
    if cmax == 0:
        s = 0
    else:
        s = (diff/cmax) * scaleFactor
    v = cmax * scaleFactor   
    
    return h, s, v  
        
            
    
```

```python
print(f_rgb_to_hsv(100,200,50,100))
```

```python
img = cv2.imread(r'colorful-flower.png')
plt.imshow(img[:,:,::-1])
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
lr = np.array([0,120,70])
ur = np.array([15,255,255])
mask1 = cv2.inRange(hsv, lr, ur)

mask1.max()
```

```python
mask1.min()
```

```python
# extract yellow color
lr = np.array([20,100,100])
ur = np.array([30,255,255])
yellow_mask = cv2.inRange(hsv, lr, ur)
res = cv2.bitwise_and(img, img, mask=yellow_mask )
```

```python
plt.figure(1)
plt.subplot(121)
plt.imshow(img[:,:,::-1])
plt.subplot(122)
plt.imshow(res[:,:,::-1])
plt.show()
```
